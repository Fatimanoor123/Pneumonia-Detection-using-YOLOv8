```python
# Install ultralytics if not already installed
!pip install ultralytics --quiet

from ultralytics import YOLO
import os

# Load your trained YOLOv8 model
model = YOLO("/content/drive/MyDrive/AI ML Projects/Healthcare/Pneumonea Detection /best.pt")  # make sure best.pt is in your Colab working directory

# Create a folder to save outputs if it doesn't exist
output_dir = "runs/detect"
os.makedirs(output_dir, exist_ok=True)

# Option 1: Run inference on a single image
single_image = "/content/drive/MyDrive/AI ML Projects/Healthcare/Pneumonea Detection /sample images/gr1.jpeg"  # replace with your image path
results = model.predict(source=single_image, save=True)
print(f"Single image inference completed. Check '{output_dir}' folder for results.")

# Option 2: Run inference on all images in a folder
folder_path = "/content/drive/MyDrive/AI ML Projects/Healthcare/Pneumonea Detection /sample images"
results = model.predict(source=folder_path, save=True, conf=0.25, imgsz=640)
print(f"Inference completed. Check 'runs/detect/' folder for results.")


# View results (Colab-friendly)
from IPython.display import Image, display
output_dir = "/content/runs/classify/predict3"

# Display the first output image
output_image = os.path.join(output_dir, os.listdir(output_dir)[0])
display(Image(output_image))
```
